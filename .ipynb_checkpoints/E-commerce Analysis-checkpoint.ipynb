```python
# %matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the main data with more than 500000 rows
main_data = pd.read_excel('./Online Retail.xlsx')
```

```python
main_data.head(10)
```

```python
main_data.tail(10)
```

```python
# check the shape of the dateframe
main_data.shape
```

```python
# check the data types in the dateframe
main_data.dtypes
```

```python
# check the datetime format
main_data.head(10)['InvoiceDate']
```

```python
# fist 6 monts of sales in 2011 
h1_2011 = main_data[(main_data['InvoiceDate'] >= '2011-01-01 00:00:00') & (main_data['InvoiceDate'] < '2011-07-01 00:00:00')]
h1_2011
```

```python
# second half 6 monts of sales in 2011
h2_2011 = main_data[(main_data['InvoiceDate'] >= '2011-07-01 00:00:00') & (main_data['InvoiceDate'] < '2011-12-31 00:00:00')]
h2_2011

#  if you want to extract this file as a new excel file, just uncomment the following and re-run the cell
# h2_2011.to_excel('Sales_second_half_of_2011.xlsx')
```

```python
canceled_h1 = []
sold_h1 = []
for record in h1_2011['InvoiceNo']:
    if 'C' in str(record):
        canceled_h1.append(record)
    else:
        sold_h1.append(record)
```

```python
# count the canceled orders during the first half of 2011
sum_canceled_h1 = len(canceled_h1)
sum_canceled_h1
```

```python
# Count the sales during the first half of 2011
sum_sales_h1 = len(sold_h1)
sum_sales_h1
```

```python
# Pie Chart of Sales vs Canceled during the first half of 2011
labels = 'Sales', 'Canceled'
sizes = [sum_sales_h1, sum_canceled_h1]
fig1, ax1 = plt.subplots()

ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
canceled_h2 = []
sold_h2 = []
for record in h2_2011['InvoiceNo']:
    if 'C' in str(record):
        canceled_h2.append(record)
    else:
        sold_h2.append(record)
```

```python
# count the canceled orders during the second half of 2011
sum_canceled_h2 = len(canceled_h2)
sum_canceled_h2
```

```python
# Count the sales during the second half of 2011
sum_sales_h2 = len(sold_h2)
sum_sales_h2
```

```python
# Pie Chart of Sales vs Canceled during the second half of 2011
labels = 'Sales', 'Canceled'
sizes = [sum_sales_h2, sum_canceled_h2]
fig1, ax1 = plt.subplots()

ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Pie Chart of Sales during the first half of 2011 vs Sales during the second half of 2011

labels = 'Sales H1', 'Sales H2'
sizes = [sum_sales_h1, sum_sales_h2]
fig1, ax1 = plt.subplots()

ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
